```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from skimage import io
```

```python
# Load the image
image = io.imread('https://upload.wikimedia.org/wikipedia/commons/thumb/4/47/PNG_transparency_demonstration_1.png/800px-PNG_transparency_demonstration_1.png')
plt.imshow(image)
plt.axis('off')  # Hide axes
plt.show()
```

```python
# Reshape the image into a 2D array of pixels
pixels = image.reshape(-1, 3)
pixels.shape  # (number_of_pixels, 3)
```

```python
# Applying K-Means with 16 clusters (colors)
kmeans = KMeans(n_clusters=16, random_state=42)
kmeans.fit(pixels)

# Replace each pixel's color with its closest cluster's centroid color
compressed_pixels = kmeans.cluster_centers_[kmeans.labels_].astype(int)
compressed_image = compressed_pixels.reshape(image.shape)
```

```python
# Display the compressed image
plt.imshow(compressed_image)
plt.axis('off')
plt.show()
```

```python
# Plot original and compressed images side by side
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Original image
ax[0].imshow(image)
ax[0].set_title('Original Image')
ax[0].axis('off')

# Compressed image
ax[1].imshow(compressed_image)
ax[1].set_title('Compressed Image (16 colors)')
ax[1].axis('off')

plt.show()
```
